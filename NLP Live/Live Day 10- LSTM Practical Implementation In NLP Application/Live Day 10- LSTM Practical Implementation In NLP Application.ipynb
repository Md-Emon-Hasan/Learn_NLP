```python
from tensorflow import keras
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.preprocessing.text import one_hot
from keras.layers import Embedding
from keras.layers import LSTM
from keras.layers import Dense
```

```python
# vocabulary size
voc_size = 5000
```

__Onehot Representation__

```python
messages = x.copy()
```

```python
messages['title']
```

```python
messages['title'][1]
```

```python
messages
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# dataset preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
corpus
```

```python
corpus[1]
```

```python
onehot_repr = [one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
onehot_repr[1]
```

__Embedding Representation__

```python
set_length = 20
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=set_length)
embedded_docs
```

```python
embedded_docs[0]
```

```python
embedded_docs[1]
```

__Embedding Representation__

```python
embedded_docs[1]
```

```python
# creating model
embedding_vector_features = 40 # features representation
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=set_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
len(embedded_docs)
```

```python
y.shape
```

```python
import numpy as np
```

```python
x_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
x_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_final,y_final,test_size=0.33,random_state=42)
```

__Model Training__

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=42)
```

__Performance Metrics And Accuracy__

```python
y_pred = model.predict(x_test)
```

```python
y_pred = np.where(y_pred > 0.5,1,0) # AUC ROC Curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
